```python
import matplotlib.pyplot as plt
import numpy as np

# Veriler
models = ["XGBoost", "RF", "LGBM", "CatBoost"]
test_mae = [9.27, 11.35, 9.10, 7.95]
test_rmse = [13.36, 15.82, 13.13, 11.76]
mape = [21.54, 27.95, 21.05, 18.12]





x = np.arange(len(models))
width = 0.25

fig, ax1 = plt.subplots(figsize=(12, 6))

# Sol eksen - MAE ve RMSE
bars1 = ax1.bar(x - width, test_mae, width, label='Test MAE', color='lightgreen')
bars2 = ax1.bar(x, test_rmse, width, label='Test RMSE', color='skyblue')
ax1.set_ylabel('MAE / RMSE')
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=45)
ax1.set_xlabel("Models")
ax1.grid(True, axis='y', linestyle='--', alpha=0.7)

# Sa eksen - MAPE
ax2 = ax1.twinx()
bars3 = ax2.bar(x + width, mape, width, label='MAPE (%)', color='pink')
ax2.set_ylabel('MAPE (%)')

# ubuk st deer yazm
def autolabel(bars, axis, decimals=2, is_percent=False):
    for bar in bars:
        height = bar.get_height()
        label = f'{height:.{decimals}f}%' if is_percent else f'{height:.{decimals}f}'
        axis.annotate(label,
                      xy=(bar.get_x() + bar.get_width() / 2, height),
                      xytext=(0, 3),
                      textcoords="offset points",
                      ha='center', va='bottom')

autolabel(bars1, ax1)
autolabel(bars2, ax1)
autolabel(bars3, ax2, is_percent=True)

# Legend' sa st keye ve ieriye yerletir
fig.legend(loc='upper right', bbox_to_anchor=(0.975, 0.98), bbox_transform=ax1.transAxes)

plt.title("NOX  Test MAE, RMSE and MAPE")
plt.tight_layout()
plt.show()
```
