```python
import os
import json
import pandas as pd


data_folder = "."  # aynı klasör
file_list = [f for f in os.listdir(data_folder) if f.endswith(".txt")]


all_records = []

# Dosyaları sırayla işle
for filename in sorted(file_list):
    file_path = os.path.join(data_folder, filename)
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    for day in data.get("days", []):
        for hour in day.get("hours", []):
            hour_record = hour.copy()
            hour_record["date"] = day["datetime"]
            all_records.append(hour_record)


weather_df = pd.DataFrame(all_records)

# Tarih + saat birleştir
weather_df["datetime"] = pd.to_datetime(weather_df["date"] + " " + weather_df["datetime"])
weather_df.drop(columns=["date"], inplace=True)

# Gereksiz sütunları çıkar
weather_df.drop(columns=['datetimeEpoch', 'stations', 'source','preciptype'], errors='ignore', inplace=True)


weather_df.sort_values("datetime", inplace=True)
weather_df.reset_index(drop=True, inplace=True)


print(weather_df.head())
print(weather_df["datetime"].min(), "→", weather_df["datetime"].max())
```

```python
import pandas as pd
import os


files = {
    "Yenibosna": "Yenibosna_istasyon.xlsx",
    "Şirinevler": "Şirinevler_istasyon.xlsx",
    "Bağcılar": "Bağcılar_istasyon.xlsx",
    "Sultangazi": "Sultangazi_istasyon.xlsx",
    "Esenyurt": "Esenyurt_istasyon.xlsx"
}

# Hedef kolon adları 
target_cols = ['datetime', 'PM10', 'SO2', 'CO', 'NO2', 'NOX', 'NO', 'O3']

all_dfs = []

for istasyon_adi, file in files.items():
    df = pd.read_excel(file)
    
    # İlk satırıu sil
    df = df.iloc[1:].reset_index(drop=True)
    df.columns = target_cols[:len(df.columns)]
    
    
    df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
    df['datetime'] = df['datetime'].dt.floor('H')  # Saatlik yuvarla

    
    for col in df.columns:
        if col != "datetime":
            df[col] = df[col].astype(str).str.replace('.', '', regex=False)
            df[col] = df[col].str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    
    df['istasyon'] = istasyon_adi
    all_dfs.append(df)

# Tüm istasyonları birleştir
df_all = pd.concat(all_dfs, ignore_index=True)

print(df_all.head())
```

```python
display(df_all.head())
df_all.shape
```

```python
kirleticiler = ['PM10', 'SO2', 'CO', 'NO2', 'NOX', 'NO', 'O3']

# Saatlik ortalama: tüm istasyonlardan gelen değerlerin ortalaması
df_avg = df_all.groupby("datetime")[kirleticiler].mean().reset_index()

print(df_avg.head())
print(df_avg.shape)
```
